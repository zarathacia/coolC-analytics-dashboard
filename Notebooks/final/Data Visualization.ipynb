```python
data="/content/drive/MyDrive/P2M/Data/Data.csv"
df = pd.read_csv(data)
df.head()
```

# Basic description

```python
df.describe()
```

Skewness

```python
df.skew(axis = 0, skipna = True)
```

Data formatting

```python
df1=df
df1['Timestamp']=df['Date']+" "+df['Time']
df1.columns
```

```python
#removal of under 0 temp values (tunis)
indexNames = df[ df['Temp'] <0 ].index
df.drop(indexNames , inplace=True)
```

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"])
df["Date"] = pd.to_datetime(df["Date"])
df['Time']=pd.to_timedelta(df['Time'])
dfm = df.groupby(['Date']).mean()
#removal of weekends
s=dfm[dfm.index.dayofweek<5]
```

Min Values

```python
dfmin = df.groupby(['Date']).min()
dfmin.head()
```

Max Values

```python
dfmax = df.groupby(['Date']).max()
dfmax.head()
```

Correlation

```python
fig = px.imshow(df.corr(), text_auto=True, aspect="auto")
fig.show()
```

# Temp study

```python
fig = px.line(df.head(500000), x='Timestamp', y='Temp', title='Temp by date first 500k entries')
fig.show()
```

```python
#Temperature Fluctuation in a day of summer
fig = px.line(df[df['Date']=="2021-08-10"], x='Timestamp', y='Temp', title='Temp by moment day of summer 2021-08-10')
fig.show()
```

```python
#Temperature Fluctuation in a day of fall
fig = px.line(df[df['Date']=="2021-11-30"], x='Timestamp', y='Temp', title='Temp by moment day of fall 2021-10-31')
fig.show()
```

```python
#Temperature Fluctuation in a day of winter
fig = px.line(df[df['Date']=="2021-12-30"], x='Timestamp', y='Temp', title='Temp by moment day of winter 2021-12-31')
fig.show()
```

```python
fig = px.histogram(df, x="Temp")
fig.update_layout(bargap=0.2)
fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=dfm.index, y=dfm['Temp'], name='Average temp by day',
                         line=dict(color='firebrick')))
fig.update_layout(title='Mean Temperature by date',
                   xaxis_title='Date',
                   yaxis_title='Temperature (degrees C°)')


fig.show()
```

Ext Temp and Temp comparaison

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=dfm.index, y=dfm['Temp'], name='Temp by iot',
                         line=dict(color='firebrick')))
fig.add_trace(go.Scatter(x=dfm.index, y=dfm['TempExt'], name=' Exterior Temp ',
                         line=dict(color='green')))
fig.update_layout(title='Mean Temperature by date',
                   xaxis_title='Date',
                   yaxis_title='Temperature (degrees C°)')


fig.show()
```

Max Temp Min Temp diff and average

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=dfmin.index, y=dfmin['Temp'], name='Min Temp',
                         line=dict(color='royalblue')))
fig.add_trace(go.Scatter(x=dfmax.index, y=dfmax['Temp'], name='Max Temp ',
                         line=dict(color='firebrick')))
fig.add_trace(go.Scatter(x=dfm.index, y=dfm['Temp'], name='Average Temp ',
                         line=dict(color='green')))
fig.update_layout(title='Mean Temperature by date',
                   xaxis_title='Date',
                   yaxis_title='Temperature (degrees C°)')


fig.show()
```
